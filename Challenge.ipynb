```python
col = ['B_30', 'B_38', 'D_114', 'D_116', 'D_117', 'D_120', 'D_126', 'D_66', 'D_68']
def cat_reducer(df, col):
    for m in col:
        x = df[m].mode()
        x = list(x)[0]
        df[m] = df[m].fillna(x).astype('int8')
cat_reducer(df, col)
```

```python
df.info()
```

```python
df.B_30.dtype
```

```python
for m in df:
    if df[m].dtype == 'float64':
        df[m] = df[m].astype('float32')
```

```python
df.info()
```

```python
df['B_31'] = df['B_31'].astype('int8')
```

```python
df.info()
```

```python
for m in df:
    if df[m].dtype == 'float32':
        df[m] = df[m].astype('float16')
```
